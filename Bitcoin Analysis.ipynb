## Bitcoin Analysis

Code to download Bitcoin Data from https://coinmarketcap.com/

```python
from cryptocmd import CmcScraper

# initialise scraper with time interval
scraper = CmcScraper("BTC", "01-01-2018", "10-03-2021")

# get raw data as list of list
headers, data = scraper.get_data()

# get data in a json format
json_data = scraper.get_data("json")

# get dataframe for the data
df = scraper.get_dataframe()
df
```

```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime

fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.update_layout(
    title='Candlestick chart for BTC',
    yaxis_title='Bitcoin Price')


fig.show()
```

```python
import plotly.graph_objects as go


from pandas_datareader import data as pdr

# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=list(df.Date), y=list(df.Close)))

# Set title
fig.update_layout(
    title_text="Bitcoin Data Trend"
)

# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show()
```
